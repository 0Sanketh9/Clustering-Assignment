```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import sklearn as sklearn
```

```python
crime=pd.read_csv("crime_data.csv")
crime
```

```python
#Hclustering
```

```python
def norm_func(i):                #Normalization or can also use MinMaxscaler   
    x = (i-i.min())/(i.max()-i.min())        
    return (x)
```

```python
df_norm = norm_func(crime.iloc[:,1:])
df_norm
```

```python
from scipy.cluster.hierarchy import linkage 
import scipy.cluster.hierarchy as sch # for creating dendrogram 
#p = np.array(df_norm) # converting into numpy array format 
z = linkage(df_norm, method="complete",metric="euclidean")
plt.figure(figsize=(15, 5))
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Index')
plt.ylabel('Distance')
sch.dendrogram(
    z,
    #leaf_rotation=0.,  # rotates the x axis labels
    #leaf_font_size=8.,  # font size for the x axis labels
)
plt.show() 
```

```python
from sklearn.cluster import AgglomerativeClustering 
h_complete = AgglomerativeClustering(n_clusters=5, linkage='complete',affinity = "euclidean").fit(df_norm) 

cluster_labels=pd.Series(h_complete.labels_)
cluster_labels
crime['clust']=cluster_labels # creating a  new column and assigning it to new column 
crime
```

```python
crime.iloc[:,1:].groupby(crime.clust).mean()
```

```python
data=crime[(crime.clust==1)]
data                               #Can Know which city belongs to which cluster depending on crime rate
```

```python
from sklearn.cluster import KMeans
```

```python
from sklearn.cluster import KMeans
fig = plt.figure(figsize=(10, 8))
WCSS = []
for i in range(1, 11):
    clf = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=0)
    clf.fit(df_norm)
    WCSS.append(clf.inertia_) # inertia is another name for WCSS

plt.plot(range(1, 11), WCSS)
plt.title('The Elbow Method')
plt.ylabel('WCSS')
plt.xlabel('Number of Clusters')
plt.show()
```

```python
clf = KMeans(n_clusters=5, init='k-means++', max_iter=300, n_init=10,  random_state=0)
y_kmeans = clf.fit_predict(df_norm)
```

```python
#DBSCAN
```

```python
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
```

```python
crimed=pd.read_csv("crime_data.csv")
crimed
```

```python
crimed.info()
```

```python
crime1 = crimed.drop(['Unnamed: 0'],axis=1) 
crime1
```

```python
array=crime1.values
array
```

```python
stscaler = StandardScaler().fit(array)
X = stscaler.transform(array) 
X 
```

```python
dbscan = DBSCAN(eps=1.3, min_samples=5)
dbscan.fit(X)   
```

```python
dbscan.labels_ 
```

```python
cl=pd.DataFrame(dbscan.labels_,columns=['cluster']) 
```

```python
cl
pd.set_option("display.max_rows", None)
cl
```

```python
crime1 = pd.concat([crimed,cl],axis=1) 
crime1
```

```python
import matplotlib.pyplot as plt
>>> plt.style.use('classic')
```

```python
crime1.plot(x="UrbanPop",y ="cluster",c=dbscan.labels_ ,kind="scatter",s=50 ,cmap=plt.cm.copper_r) 
plt.title('Clusters using DBScan') 
```

```python
dl = dbscan.labels_ 
```

```python
sklearn.metrics.silhouette_score(X, dl)
```
